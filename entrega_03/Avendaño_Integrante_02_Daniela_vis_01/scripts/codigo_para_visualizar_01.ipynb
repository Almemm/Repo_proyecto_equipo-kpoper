```python
# Primero importar el archivo csv de la base de datos a utilizar

import pandas as pd
df1 = pd.read_csv('/content/conciertos_database_ocupado.csv.csv', delimiter=',')
df1
```

```python
# Crear dataframe para la visualización que se quiere

df1_mercado = df1.groupby(["GRUPO", "MERCADO"]).agg({"PAÍS": "count"}).reset_index()
df1_mercado
```

```python
# Crear gráfico de barras que conteste: ¿Qué mercado es al que más apuntan los grupos según la cantidad de conciertos que han realizado en el lugar?
# Al principio se duplicaban ciertas opciones, por lo que primero se deja un código para que no se produzca el error.

df1_mercado['MERCADO'] = (
    df1_mercado['MERCADO']
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
    .str.upper())

import altair as alt

chart = alt.Chart(df1_mercado).mark_bar().encode(
    x=alt.X('PAÍS:Q', title='Cantidad de conciertos'),
    y=alt.Y('MERCADO:N', title='Mercado'),
    color=alt.Color('GRUPO:N', title='GRUPOS', scale=alt.Scale(scheme='tableau10')),
    tooltip=[
        alt.Tooltip('GRUPO:N', title='Grupo'),
        alt.Tooltip('PAÍS:Q', title='Cantidad de conciertos'),
        alt.Tooltip('MERCADO:N', title='Mercado')
    ]
).properties(
    width=1000,
    height=500,
    title='Cantidad de conciertos por mercado'
).interactive()

chart
```

```python
# Exportar la visualización:

chart.save('vis_01.html')
```
