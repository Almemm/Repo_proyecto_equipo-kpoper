```python
import numpy as np

def limpiar(col):
    def limpiar_valor(x):
        if pd.isna(x):
            return np.nan
        # Convertir a string y limpiar
        s = str(x).strip()
        # Eliminar todo excepto dígitos y punto decimal
        s = ''.join(ch for ch in s if ch.isdigit() or ch == '.')
        # Si queda vacío, devolver NaN
        if s == '':
            return np.nan
        # Si hay más de un punto decimal, devolver NaN (número inválido)
        if s.count('.') > 1:
            return np.nan
        try:
            return float(s)
        except:
            return np.nan

    return col.apply(limpiar_valor)

  # Supongamos que tienes un DataFrame df con las columnas originales
df["ENTRADA MÁS BARATA"] = limpiar(df["ENTRADA MÁS BARATA"])
df["ENTRADA MÁS CARA"] = limpiar(df["ENTRADA MÁS CARA"])

df
```
