```python
import pandas as pd
import altair as alt

# Cargar la base de datos
GEN = pd.read_csv("/content/Grupos_KPOP.csv", delimiter=";")

# Filtrar los datos entre 1990 y 2025 seg煤n el a帽o de debut
GEN = GEN[(GEN["AO_DEBUT"] >= 1990) & (GEN["AO_DEBUT"] <= 2025)]

# Agrupar y contar los grupos por a帽o de debut y g茅nero
GEN_grouped = GEN.groupby(["AO_DEBUT", "GENERO"]).size().reset_index(name="Cantidad")

# Crear la selecci贸n de intervalo (Brush)
brush = alt.selection_interval(encodings=["x"])

# Crear la selecci贸n del mouse (Hover)
highlight = alt.selection_single(on="mouseover", fields=["AO_DEBUT"], empty="none")

# Opacidad basada en la selecci贸n del brush
opacity_brush = alt.condition(brush, alt.value(0.9), alt.value(0.1))

# Opacidad basada en la selecci贸n de hover
opacity_highlight = alt.condition(highlight, alt.value(1), alt.value(0.3))

# 娥拣答筐吼 con interactividad dual
overview = alt.Chart(GEN_grouped).mark_bar().encode(
    alt.X("AO_DEBUT:O", axis=alt.Axis(title="A帽o de debut")),
    alt.Y("Cantidad:Q", title="Cantidad de grupos"),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="G茅nero")),
    opacity=opacity_brush,  # Opacidad basada en el brush
    tooltip=["AO_DEBUT", "Cantidad", "GENERO"]  # Muestra informaci贸n detallada al pasar el cursor
).add_selection(brush).add_selection(highlight)  # Agregar ambas selecciones

# 拆桂桂 de puntos con resaltado din谩mico
detail = alt.Chart(GEN_grouped).mark_point().encode(
    alt.X("AO_DEBUT:O", axis=alt.Axis(title="A帽o de debut")),
    alt.Y("Cantidad:Q", axis=alt.Axis(title="Cantidad de Grupos")),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="G茅nero")),
    opacity=opacity_highlight  # Opacidad basada en el hover
) .properties(title="Distribuci贸n de grupos por a帽o")

# Concatenar gr谩ficos
graf = overview & detail
graf

```
