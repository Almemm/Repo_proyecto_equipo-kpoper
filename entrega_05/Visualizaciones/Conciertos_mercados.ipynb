```python

import pandas as pd
df1 = pd.read_csv('/content/BASE DE DATOS DE CONCIERTOS ARREGLADO.csv', delimiter=',')
df1
```

```python
df1_mercado = df1.groupby(["GRUPO", "MERCADO"]).agg({"PAÍS": "count"}).reset_index()
df1_mercado
```

```python
df1_mercado['MERCADO'] = (
    df1_mercado['MERCADO']
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
    .str.upper())


import altair as alt

# Define un diccionario de colores por grupo
color_map = {
    'ARTMS': '#d1cfe2',
    'BTS': '#D460FF',
    "Girls' Generation (SNSD)": '#ffb1c1',
    'I-DLE': '#672e78',
    'SEVENTEEN': '#9cadce',
    'SHINee': '#4BEDF3',
    'Stray Kids': '#EB8484',
    'Super Junior': '#008EB4',
    'TWICE': '#FFE281',
    'aespa': '#a9daff'
}


# Crea una escala de color personalizada
color_scale = alt.Scale(domain=list(color_map.keys()), range=list(color_map.values()))

chart = alt.Chart(df1_mercado).mark_bar().encode(
    x=alt.X('PAÍS:Q', title='Cantidad de conciertos'),
    y=alt.Y('MERCADO:N', title='Mercado'),
    color=alt.Color('GRUPO:N', title='Grupos', scale=color_scale),
    tooltip=[
        alt.Tooltip('GRUPO:N', title='Grupo'),
        alt.Tooltip('PAÍS:Q', title='Cantidad de conciertos'),
        alt.Tooltip('MERCADO:N', title='Mercado')
    ]
).properties(
    width=1000,
    height=500,
    title='Cantidad de conciertos por mercado'
).interactive()


chart
```

```python
chart.save('vis_dani_cronica.html')
```
