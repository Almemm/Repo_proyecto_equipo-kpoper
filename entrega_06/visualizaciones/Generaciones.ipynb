```python
import pandas as pd
import altair as alt

GEN = pd.read_csv("/content/Grupos_KPOP.csv", delimiter=";")
GEN = GEN[(GEN["AÑO_DEBUT"] >= 1990) & (GEN["AÑO_DEBUT"] <= 2025)]
GEN_grouped = GEN.groupby(["AÑO_DEBUT", "GENERO"]).size().reset_index(name="Cantidad")

def asignar_generacion(año):
    if 1990 <= año <= 2002:
        return 'Primera generación'
    elif 2003 <= año <= 2011:
        return 'Segunda generación'
    elif 2012 <= año <= 2017:
        return 'Tercera generación'
    elif 2018 <= año <= 2022:
        return 'Cuarta generación'
    elif 2023 <= año <= 2025:
        return 'Quinta generación'
    return 'Sin generación'

GEN_grouped['Generacion'] = GEN_grouped['AÑO_DEBUT'].apply(asignar_generacion)

colores_generacion = {
    'Primera generación': '#D4AFB9',
    'Segunda generación': '#D1CFE2',
    'Tercera generación': '#9CADCE',
    'Cuarta generación': '#7EC5CF',
    'Quinta generación': '#DAEAF6'
}

# Base chart con eje x más claro
base = alt.Chart(GEN_grouped).mark_bar().encode(
    x=alt.X('AÑO_DEBUT:O', title='Año de debut',
            axis=alt.Axis(labelAngle=0, labelFontSize=10)),  # Mantiene años legibles
    y=alt.Y('Cantidad:Q', title='Cantidad de grupos'),
    color=alt.Color('Generacion:N',
                    scale=alt.Scale(domain=list(colores_generacion.keys()),
                                    range=list(colores_generacion.values())),
                    legend=alt.Legend(title='Generación')),
    tooltip=['AÑO_DEBUT', 'GENERO', 'Cantidad', 'Generacion']
).properties(
    width=800,
    height=120
)

chart = base.facet(
    row=alt.Row('GENERO:N', title=None, header=alt.Header(labelFontSize=14, labelFontWeight="bold"))
).resolve_scale(
    x='independent'
).properties(
    title='Cantidad de grupos K-Pop por año y género'
)

chart
```

```python
chart.save('grafico_kpop.html')
```

```python
import pandas as pd
import altair as alt
import colorsys

# Cargar datos
GEN = pd.read_csv("/content/Grupos_KPOP_actualizada.csv", delimiter=";")
# GEN = pd.read_csv("/content/Grupos_KPOP.csv", delimiter=";")
GEN = GEN[(GEN["AÑO_DEBUT"] >= 1990) & (GEN["AÑO_DEBUT"] <= 2025)]

# Agrupar datos
GEN_grouped = GEN.groupby(["AÑO_DEBUT", "GENERO"]).size().reset_index(name="Cantidad")

# Asignar generación y etiquetas
def asignar_generacion(año):
    if 1990 <= año <= 2002:
        return 1, 'Primera generación'
    elif 2003 <= año <= 2011:
        return 2, 'Segunda generación'
    elif 2012 <= año <= 2017:
        return 3, 'Tercera generación'
    elif 2018 <= año <= 2022:
        return 4, 'Cuarta generación'
    elif 2023 <= año <= 2025:
        return 5, 'Quinta generación'
    return None, None

GEN_grouped[['Generacion', 'Etiqueta']] = GEN_grouped['AÑO_DEBUT'].apply(lambda año: pd.Series(asignar_generacion(año)))

# Ajustar luminosidad
def ajustar_luminosidad(hex_color, factor):
    hex_color = hex_color.lstrip('#')
    r, g, b = tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))
    h, l, s = colorsys.rgb_to_hls(r/255, g/255, b/255)
    l = max(0, min(1, l * factor))
    r_new, g_new, b_new = colorsys.hls_to_rgb(h, l, s)
    return '#{:02x}{:02x}{:02x}'.format(int(r_new*255), int(g_new*255), int(b_new*255))

# Colores base por generación
colores_generacion = {
    1: '#D4AFB9',
    2: '#D1CFE2',
    3: '#9CADCE',
    4: '#7EC5CF',
    5: '#DAEAF6'
}

# Factores de luminosidad por género
factores_genero = {
    'Hombres': 0.7,
    'Mixto': 0.85,
    'Mujeres': 1
}

# Asignar color final con ajuste de luminosidad
GEN_grouped['Color'] = GEN_grouped.apply(
    lambda row: ajustar_luminosidad(
        colores_generacion[row['Generacion']],
        factores_genero.get(row['GENERO'], 1.0)
    ),
    axis=1
)

# Interactividad
brush = alt.selection_interval(encodings=["x"])
highlight = alt.selection_single(on="mouseover", fields=["AÑO_DEBUT"], empty="none")

opacity_brush = alt.condition(brush, alt.value(0.9), alt.value(0.1))
opacity_highlight = alt.condition(highlight, alt.value(1), alt.value(0.3))

# Gráfico final con color personalizado
overview = alt.Chart(GEN_grouped).mark_bar().encode(
    alt.X("AÑO_DEBUT:O", axis=alt.Axis(title="Año de debut")),
    alt.Y("Cantidad:Q", title="Cantidad de grupos"),
    alt.Color("Color:N", scale=None, legend=None),  # <- usar color calculado
    alt.Order("GENERO:N", sort="ascending"),
    opacity=opacity_brush,
    tooltip=["AÑO_DEBUT", "Cantidad", "GENERO", "Etiqueta"],
).add_selection(brush).add_selection(highlight)

# Para agregar la font, debe estar importada en el html o no pasará nada, luego guardar este graf como html y después de traspasarlo se debería ver
# Use 'overview' instead of 'graf' as 'graf' is not defined at this point
overview.configure_axis(
    labelFont='Gowun Dodum',
    titleFont='Gowun Dodum'
).configure_title(
    font='Gowun Dodum'
).configure_legend(
    labelFont='Gowun Dodum',
    titleFont='Gowun Dodum'
)

# Datos de leyenda
leyenda_data = pd.DataFrame({
    'Generacion': [1, 2, 3, 4, 5],
    'Etiqueta': ['Primera generación', 'Segunda generación', 'Tercera generación', 'Cuarta generación', 'Quinta generación'],
    'Color': ['#D4AFB9', '#D1CFE2', '#9CADCE', '#7EC5CF', '#DAEAF6']
})

# Gráfico de leyenda
leyenda = alt.Chart(leyenda_data).mark_rect(width=20, height=20).encode(
    y=alt.Y('Etiqueta:N', sort=leyenda_data['Etiqueta'].tolist(), title=None, axis=None),
    color=alt.Color('Color:N', scale=None),
).properties(
    width=150,
    height=120,
    title='Colores por generación',
)

# Añadir texto al lado
etiquetas = alt.Chart(leyenda_data).mark_text(align='left', dx=25).encode(
    y=alt.Y('Etiqueta:N', sort=leyenda_data['Etiqueta'].tolist()),
    text='Etiqueta:N'
)

# Combinar leyenda color + texto
leyenda_final = leyenda + etiquetas

# Use 'overview' in the combined chart as 'graf' was not explicitly used for the main chart
chart_final = overview | leyenda_final
chart_final
```

```python
chart_final.save('grafico_kpop.html')
```
