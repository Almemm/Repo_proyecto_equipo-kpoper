```python
import pandas as pd
import altair as alt

# Cargar la base de datos
GEN = pd.read_csv("/content/Grupos_KPOP.csv", delimiter=";")

# Filtrar los datos entre 1990 y 2025 según el año de debut
GEN = GEN[(GEN["AÑO_DEBUT"] >= 1990) & (GEN["AÑO_DEBUT"] <= 2025)]

# Agrupar y contar los grupos por año de debut y género
GEN_grouped = GEN.groupby(["AÑO_DEBUT", "GENERO"]).size().reset_index(name="Cantidad")

# Crear el "brush selection"
brush = alt.selection_interval(encodings=["x"])

# Crear selección del mouse (Hover)

highlight = alt.selection(
    type="single", on="mouseover", fields=["AÑO_DEBUT"], nearest=True
)

# Determinar la opacidad en base al "brush selection"
opacity = alt.condition(brush, alt.value(0.9), alt.value(0.1))

# Determinar la opacidad en la selección de hover

hightlight_opacity = alt.condition(highlight, alt.value(1), alt.value(0.5))

# Histograma de grupos por año con diferenciación por género
overview = alt.Chart(GEN_grouped).mark_bar().encode(
    alt.X("AÑO_DEBUT:O", axis=alt.Axis(title="Año de debut")),
    alt.Y("Cantidad:Q", title="Cantidad de grupos"),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="Género")),
    opacity=opacity # Changed from Opacity
).add_selection(brush).properties(height=50)

# Gráfica de dispersión mostrando cantidad de grupos por género y año de debut
detail = alt.Chart(GEN_grouped).mark_point().encode(
    alt.X("AÑO_DEBUT:O", axis=alt.Axis(title="Año de debut")),
    alt.Y("Cantidad:Q", axis=alt.Axis(title="Cantidad de Grupos")),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="Género")),
    opacity=opacity # Changed from Opacity
)

# Concatenar las gráficas
graf = overview & detail # Changed from Detail
graf
```

```python
import pandas as pd
import altair as alt

# Cargar la base de datos
GEN = pd.read_csv("/content/Grupos_KPOP.csv", delimiter=";")

# Filtrar los datos entre 1990 y 2025 según el año de debut
GEN = GEN[(GEN["AÑO_DEBUT"] >= 1990) & (GEN["AÑO_DEBUT"] <= 2025)]

# Agrupar y contar los grupos por año de debut y género
GEN_grouped = GEN.groupby(["AÑO_DEBUT", "GENERO"]).size().reset_index(name="Cantidad")

# Crear la selección de intervalo (Brush)
brush = alt.selection_interval(encodings=["x"])

# Crear la selección del mouse (Hover)
highlight = alt.selection_single(on="mouseover", fields=["AÑO_DEBUT"], empty="none")

# Opacidad basada en la selección del brush
opacity_brush = alt.condition(brush, alt.value(0.9), alt.value(0.1))

# Opacidad basada en la selección de hover
opacity_highlight = alt.condition(highlight, alt.value(1), alt.value(0.3))

# 𝗛𝗶𝘀𝘁𝗼𝗴𝗿𝗮𝗺𝗮 con interactividad dual
overview = alt.Chart(GEN_grouped).mark_bar().encode(
    alt.X("AÑO_DEBUT:O", axis=alt.Axis(title="Año de debut")),
    alt.Y("Cantidad:Q", title="Cantidad de grupos"),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="Género")),
    opacity=opacity_brush,  # Opacidad basada en el brush
    tooltip=["AÑO_DEBUT", "Cantidad", "GENERO"]  # Muestra información detallada al pasar el cursor
).add_selection(brush).add_selection(highlight)  # Agregar ambas selecciones

# 𝗗𝗲𝘁𝗮𝗹𝗹𝗲 de puntos con resaltado dinámico
detail = alt.Chart(GEN_grouped).mark_point().encode(
    alt.X("AÑO_DEBUT:O", axis=alt.Axis(title="Año de debut")),
    alt.Y("Cantidad:Q", axis=alt.Axis(title="Cantidad de Grupos")),
    alt.Color("GENERO:N", scale=alt.Scale(scheme="category10"), legend=alt.Legend(title="Género")),
    opacity=opacity_highlight  # Opacidad basada en el hover
) .properties(title="Distribución de grupos por año")

# Concatenar gráficos
graf = overview & detail
graf

```
